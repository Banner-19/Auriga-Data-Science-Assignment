```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
```

# Load example image

```python
image_path = 'D:\Internship Assignments\Auriga Tech Solutions Private Limited\images.jpeg'  # Replace with your image path
image = Image.open(image_path).convert('RGB')
image = image.resize((224, 224))  # Resize for consistency
image_array = np.array(image)
```

# Define a simple edge-detection filter (3x3 Sobel filter)

```python
sobel_filter = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])
```

# Apply the filter using convolution

## Since image_array is RGB, we apply the filter to each channel separately

```python
feature_maps = np.stack([convolve(image_array[:, :, i], sobel_filter) for i in range(3)], axis=2)
```

## Normalize the feature maps to display

```python
feature_maps = (feature_maps - feature_maps.min()) / (feature_maps.max() - feature_maps.min())
```

# Display original image and feature map

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title('Original Image')
axes[1].imshow(feature_maps)
axes[1].set_title('Feature Map (Edge Detection)')

plt.show()
```
